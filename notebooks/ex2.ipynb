# Ejercicio 2 - LPF y Threshold

```python
import cv2 as cv
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

### Carga de la imagen

```python
img = cv.imread('../resources/stars.png', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
```

## Low-pass filter
Se utiliza un filtro promediador de 15x15.

```python
n = 15
kernel = np.ones((n,n), np.float32) / n**2
filtered_img = cv.filter2D(img, -1, kernel)

plt.imshow(filtered_img, cmap='gray')
plt.axis('off')
plt.show()
```

## Threshold Application
Se aplica un threshold de 25% para obtener los objetos ms representativos de la imagen.

```python
th = 0.25 * 255
_, thresholded_img = cv.threshold(filtered_img, th, 255, cv.THRESH_BINARY)
plt.imshow(thresholded_img, cmap='gray')
plt.axis('off')
plt.show()
```

## Resultado final

```python
plt.figure(figsize=(15,8)) 
ax = plt.subplot(1,3,1)
ax.imshow(img, cmap='gray')
ax.axis('off')
ax.set_title('Original')

ax = plt.subplot(1,3,2)
ax.imshow(filtered_img, cmap='gray')
ax.set_title(f'Low Pass {n}x{n}')
ax.axis('off')

ax = plt.subplot(1,3,3)
ax.imshow(thresholded_img, cmap='gray')
ax.set_title(f'Threshold {th / 255 * 100:.0f}%')
ax.axis('off')

plt.show()
```

De esta forma, se pueden detectar los objetos ms representativos de la imagen. Subiendo el threshold se lo puede hacer ms selectivo.
