```python
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
import numpy as np
```

```python
# Read the image from the file system
image = cv2.imread('../resources/cactus.png', cv2.IMREAD_GRAYSCALE)
image_dark = cv2.imread('../resources/cactus_dark.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Plot the image
plt.imshow(image, cmap='gray')
plt.show()
plt.imshow(image_dark, cmap='gray')
plt.show()
```

```python
# Function to apply High Boost filter including diagonals
def diagHighBoostFilter(image, A=1):
    kernel = np.array([[-1, -1, -1], 
                        [-1, A + 8, -1],
                        [-1, -1, -1]])
    return cv2.filter2D(image, ddepth=-1, kernel=kernel)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0][0].imshow(image_dark, cmap='gray')
ax[0][0].set_title('Original')
ax[0][1].imshow(diagHighBoostFilter(image_dark, 0), cmap='gray')
ax[0][1].set_title('A=0')
ax[1][0].imshow(diagHighBoostFilter(image_dark, 1), cmap='gray')
ax[1][0].set_title('A=1')
ax[1][1].imshow(diagHighBoostFilter(image_dark, 1.7), cmap='gray')
ax[1][1].set_title('A=1.7')

plt.show()
```

```python
# Function to apply High Boost filter including diagonals with another kernel
def diagHighBoostFilterModified(image, A=1):
    kernel = np.array([[-1, -1, -1], 
                        [-1, 9*A - 1, -1],
                        [-1, -1, -1]])
    return cv2.filter2D(image, ddepth=-1, kernel=kernel*1/9)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0][0].imshow(image, cmap='gray')
ax[0][0].set_title('Original')
ax[0][1].imshow(diagHighBoostFilterModified(image, 1), cmap='gray')
ax[0][1].set_title('A=1')
ax[1][0].imshow(diagHighBoostFilterModified(image, 1.5), cmap='gray')
ax[1][0].set_title('A=1.5')
ax[1][1].imshow(diagHighBoostFilterModified(image, 1.7), cmap='gray')
ax[1][1].set_title('A=1.7')

plt.show()
```

# Anlisis
Al convolucionar la imagen por un kernel con A > 1 lo que esta sucediendo es que se estan potenciando las componentes de alta frecuencia de la imgen y por eso se aprecia un efecto de sharpening en los bordes de la imgen.
